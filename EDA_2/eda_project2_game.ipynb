### 1 Which genres have the highest average user ratings?

```python
# 1 Which genres have the highest average user ratings?
genre_ratings=df.groupby("Genre")["User_Rating"].mean().sort_values(ascending=False)
print(genre_ratings)

plt.figure(figsize=(12, 6))
sns.barplot(x=genre_ratings.index, y=genre_ratings.values, palette='viridis',edgecolor='black')
plt.xticks(rotation=45)
plt.title('Average User Rating by Genre')
plt.ylabel('Average Rating')
plt.xlabel('Genre')
plt.show()
```

### 2 How has the number of game releases changed over the years?

```python
# 2 How has the number of game releases changed over the years?
release_trend = df['Release_Year'].value_counts().sort_values(ascending=False)
print(release_trend)

plt.figure(figsize=(12, 6))
sns.lineplot(x=release_trend.index, y=release_trend.values, marker='o', linestyle='-')
plt.title('Number of Games Released Over the Years',fontsize=16,fontweight='bold')
plt.xlabel('Year',fontsize=14)
plt.ylabel('Number of Games',fontsize=14)
plt.grid(True)
plt.show()

```

### 3 which age group plays the game most

```python
# 3 which age group plays the game most
most_common_age_group = df["Age_Group_Targeted"].value_counts().sort_values(ascending=False)
print(most_common_age_group)

df["Age_Group_Targeted"].value_counts().plot(kind="bar", figsize=(8,5), color="skyblue", edgecolor="black")
plt.title("Number of Games by Age Group",fontsize=16,fontweight='bold')
plt.xlabel("Age Group",fontsize=14)
plt.ylabel("Count",fontsize=14)
plt.xticks(rotation=0)
plt.show()
```

 ### 4 top 5 developers with the highest average user rating?

### 15 which publisher release the longest game?

```python
# 15 which publisher release the longest game?

longest_game_publisher = df.groupby("Publisher")["Game_Length_(Hours)"].mean().idxmax()
max_avg_length = df.groupby("Publisher")["Game_Length_(Hours)"].mean().max()

print(f"The publisher with the longest average game length is: {longest_game_publisher} ({max_avg_length:.2f} hours)")

# Top 10 publishers with longest games
top_publishers = df.groupby("Publisher")["Game_Length_(Hours)"].mean().nlargest(10)

# Visualization
plt.figure(figsize=(12, 6))
sns.barplot(y=top_publishers.index, x=top_publishers.values, palette="Blues_r",edgecolor='black')
plt.xlabel("Average Game Length (Hours)",fontsize=14)
plt.ylabel("Publisher",fontsize=14)
plt.title("Top 10 Publishers with Longest Average Game Length",fontsize=16,fontweight='bold')
plt.show()
```

# suggestions

1. Focus on High-Rated Genres for Development & Marketing
Party, Fighting, and RPG games have the highest user ratings, suggesting strong player engagement in these genres.
Developers should invest in these genres to maximize user satisfaction and potentially higher sales.


2. Capitalize on the Most Popular Genres
RPGs, Shooters, and Strategy games are the most played genres.
Developers should focus on these categories when designing new games, as they have an established audience and market demand.


3. Track the Decline in Game Releases
The number of game releases peaked in 2020 but has declined in 2023.
This could be due to market saturation, development delays, or shifting industry trends (e.g., focus on quality over quantity).
Developers and publishers should assess why releases have dropped and adapt their strategies accordingly.


4. Target the Right Age Groups
Teens and Kids play the most games, followed closely by All Ages and Adults.
Developers can create age-appropriate content and marketing campaigns to appeal to these groups.
Multiplayer and competitive games may appeal more to teens, while immersive single-player games might work better for adults.


5. Leverage the Most Trusted Game Developers
Nintendo, Epic Games, EA Sports, Capcom, and Valve have the highest-rated games.
Collaborating with or learning from these companies can help improve game quality and market success.


6. Price Optimization Strategy
The most expensive games ($59.99) include well-known franchises like Counter-Strike, Halo, and Tomb Raider.
Players are willing to pay premium prices for high-quality, established titles, meaning new game developers should focus on quality, franchise-building, and strong brand loyalty.
Offering DLCs, expansions, and premium versions could also increase revenue.





### conclusion

The gaming industry remains strong, with high demand for RPGs, Shooters, and Strategy games.


Multiplayer gaming is highly relevant, and developers should consider cross-platform compatibility.


Nintendo, Epic Games, and other top-rated developers set quality benchmarks that new studios should follow.


The decline in new game releases in 2023 suggests a shift in industry trends, possibly focusing on quality over quantity

Valve's long-form game model works well, but only for certain genres.


RPGs, Adventure, and Strategy games should prioritize length, while shorter genres like Fighting or Party games should focus on replayability.


Developers should not force length but instead build depth and engagement.


A hybrid approach of offline and online features can enhance game longevity and player retention.
