### Data Analysis And Ploting


```python
# calculate correlation
correlation_matrix=df.corr()


# plot the heat map
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Show the plot
plt.title("Correlation Heatmap",fontsize=18,fontweight='bold')
plt.show()
```

### 1 what is the average cargo weight for each ship type?

```python

avr_cargo_weight=df.groupby('Ship_Type')['Cargo_Weight_Tons'].mean()
print(avr_cargo_weight)

# Bar Plot
plt.figure(figsize=(9,6))
avr_cargo_weight.plot(kind='bar',color='skyblue',edgecolor='black')
plt.title('Average Cargo Weight For Each Ship Type',fontsize=16,fontweight='bold')
plt.xlabel('Ship Type',fontsize=14)
plt.ylabel('Average Cargo Weight (Tons)',fontsize=14)
plt.show()
```

### 2 How Does The Weather Condition Affect The Speed

```python

avg_speed_weather=df.groupby('Weather_Condition')['Speed_Over_Ground_Knots'].mean()
print(avg_speed_weather)

# bar Plot
plt.figure(figsize=(9,6))
sns.barplot(data=df,x='Weather_Condition',y='Speed_Over_Ground_Knots',palette='Blues',edgecolor='black')
plt.title("Weather condition affect the speed",fontsize=16,fontweight='bold')
plt.xlabel('Weather Conditions',fontsize=14)
plt.ylabel('Speed Over Ground Knots',fontsize=14)
```

### 3 What is the Seasonal Impact Across Diffrent Ship Type?

```python

seasonal_impact=df.groupby('Ship_Type')['Seasonal_Impact_Score'].mean()
print(seasonal_impact)

# line plot
plt.figure(figsize=(8,6))
sns.lineplot(data=df, x='Ship_Type', y='Seasonal_Impact_Score')
plt.title('Seasonal Impact Across Diffrent Ship Type',fontsize=18,fontweight='bold')
plt.xlabel('Ship Type',fontsize=16)
plt.ylabel('Seasonal Impact Score',fontsize=16)
plt.show()
```

### 4 What is the overall average speed of all voyages?

```python


avg_speed = df.groupby('Ship_Type')['Speed_Over_Ground_Knots'].mean()
print(avg_speed)
plt.figure(figsize=(9,6))
sns.barplot(data=df,x='Ship_Type',y='Speed_Over_Ground_Knots',palette='Greens',edgecolor='black')
plt.title('Overall Average Speed Of All Voyages',fontsize=18,fontweight='bold')
plt.xlabel('Ship Type',fontsize=16)
plt.ylabel("Speed(Knocks)",fontsize=16)
plt.show()
```

### 5 Which Ship Has The Highest Operational Cost

```python

max_operational_cost=df.groupby('Ship_Type')['Operational_Cost_Usd'].mean()
print(max_operational_cost)

# bar plot
plt.figure(figsize=(9,6))
max_operational_cost.plot(kind='bar',color='yellow',edgecolor='black')
plt.title('Ship Has The Highest Operational Cost',fontsize=18,fontweight='bold')
plt.xlabel('Ship Type',fontsize=16)
plt.ylabel('Operational Cost Usd',fontsize=16)
plt.show()
```

### 6 Boxplot of Cargo Weight tons by WeatherCondition

```python

avg_cargo=df.groupby('Cargo_Weight_Tons')['Weather_Condition']
print(avg_cargo)
plt.figure(figsize=(9, 6))
sns.boxplot(data=df, x='Weather_Condition', y='Cargo_Weight_Tons', palette='viridis')
plt.title('Boxplot: Cargo Weight by Weather Condition',fontsize=18,fontweight='bold')
plt.xlabel('Weather Condition',fontsize=16)
plt.ylabel('Cargo Weight (tons)',fontsize=16)
plt.xticks(rotation=45)  # Rotate x labels for readability
plt.show()
```

### 7 What is the relationship between maintenance status and engine power?

```python

maintenance_power = df.groupby('Maintenance_Status')['Engine_Power_Kw'].mean()
print(maintenance_power)

plt.figure(figsize=(9,6))
sns.lineplot(data=df, x='Maintenance_Status', y='Engine_Power_Kw', estimator='mean')
plt.title('Average Engine Power by Maintenance Status',fontsize=18,fontweight='bold')
plt.xlabel('Maintanance Status',fontsize=16)
plt.ylabel('Engine Power',fontsize=16)
plt.show()
```

### 8 which engine type ship is more faster

```python

avg_speed_by_engine = df.groupby('Engine_Type')['Speed_Over_Ground_Knots'].mean().reset_index()

# Sort values so that the fastest engine type appears first
avg_speed_by_engine = avg_speed_by_engine.sort_values(by='Speed_Over_Ground_Knots', ascending=False)

# Print the results for inspection
print(avg_speed_by_engine)

# Create a bar plot to visualize the average speed by engine type
plt.figure(figsize=(9, 6))
sns.barplot(data=avg_speed_by_engine, x='Engine_Type', y='Speed_Over_Ground_Knots', palette='viridis',edgecolor='black')
plt.title("Average Speed Over Ground by Engine Type",fontsize=18,fontweight='bold')
plt.xlabel("Engine Type",fontsize=16)
plt.ylabel("Average Speed (knots)",fontsize=16)
plt.show()


```

### 9 How does the average load percentage change by ship type?

```python

avg_load_percentage = df.groupby('Ship_Type')['Average_Load_Percentage'].mean()
print(avg_load_percentage)

# bar plot#
plt.figure(figsize=(9,6))
sns.barplot(data=df, x='Ship_Type', y='Average_Load_Percentage', estimator='mean',palette='pastel',edgecolor='black')
plt.title('Average Load Percentage by Ship Type',fontsize=18,fontweight='bold')
plt.xlabel('Ship Type',fontsize=16)
plt.ylabel('Average Load Percentage',fontsize=16)
plt.show()
```

### 10 which type ship travels more distance?

```python

total_distance_by_ship = df.groupby('Ship_Type')['Distance_Traveled_Nm'].sum().reset_index()
print(total_distance_by_ship)

# Sort the data to highlight the ship type with the highest total distance
total_distance_by_ship = total_distance_by_ship.sort_values(by='Distance_Traveled_Nm', ascending=False)

# Plot the total distance traveled by each ship type
plt.figure(figsize=(10, 6))
sns.barplot(data=total_distance_by_ship, x='Ship_Type', y='Distance_Traveled_Nm', palette='mako',edgecolor='black')
plt.title("Total Distance Traveled (nautical miles) by Ship Type",fontsize=18,fontweight='bold')
plt.xlabel("Ship Type",fontsize=16)
plt.ylabel("Total Distance Traveled (nm)",fontsize=16)
plt.show()
```

### 11 diffrent types of ships

```python
# 11 diffrent types of ships

ship_counts = df['Ship_Type'].value_counts()


cmap = plt.get_cmap("Set3")  # Choose a colormap
colors = cmap(range(len(ship_counts))) 

# Create a pie chart
plt.figure(figsize=(10, 7))
plt.pie(
    ship_counts, 
    labels=ship_counts.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors, 
    explode=[0.05] * len(ship_counts),  # Slight explosion effect
    shadow=True, 
    wedgeprops={'edgecolor': 'black'}
)

plt.title('Distribution of Ship Types', fontsize=16, fontweight='bold')
plt.show()
```

### 12 what are the diffrent route tytpes

```python
# 12 what are the diffrent route tytpes
route_counts = df['Route_Type'].value_counts()
print(route_counts)

# Define colors
colors = plt.get_cmap("Set2")(range(len(route_counts)))

# Create the pie chart
plt.figure(figsize=(10, 7))
plt.pie(route_counts, labels=route_counts.index, autopct='%1.1f%%', 
        startangle=140, colors=colors, explode=[0.05] * len(route_counts),
        shadow=True, wedgeprops={'edgecolor': 'black'})

plt.title('Distribution of Different Route Types', fontsize=14, fontweight='bold')
plt.show()
```

### 13 Which Ship Type is Most Profitable?

```python
# 13 Which Ship Type is Most Profitable?
avg_profit=df.groupby('Ship_Type')['Revenue_Per_Voyage_Usd'].sum()
print(avg_profit)
plt.figure(figsize=(10, 6))
sns.barplot(x='Ship_Type', y='Revenue_Per_Voyage_Usd', data=df, estimator=sum, palette='viridis')
plt.xlabel('Ship Type', fontsize=16)
plt.ylabel('Total Revenue (USD)', fontsize=16)
plt.title('Total Revenue by Ship Type', fontsize=18, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 14 What are the busiest routes based on weekly voyage count?

```python
# 14 What are the busiest routes based on weekly voyage count?
plt.figure(figsize=(12, 6))
route_voyages = df.groupby("Route_Type")["Weekly_Voyage_Count"].sum().sort_values(ascending=False)
print(route_voyages)

# Plot the busiest routes
sns.barplot(x=route_voyages.index, y=route_voyages.values, palette="Blues_r",edgecolor='black')
plt.xticks(rotation=45)
plt.xlabel("Route Type",fontsize=16)
plt.ylabel("Total Weekly Voyages",fontsize=16)
plt.title("Busiest Routes Based on Weekly Voyage Count",fontsize=18,fontweight='bold')
plt.show()
```

### 15 Does Route Type Impact Revenue?

```python
# 15 Does Route Type Impact Revenue?

route_revenue = df.groupby('Route_Type')['Revenue_Per_Voyage_Usd'].sum()
print(route_revenue)

# Define colors
colors = plt.get_cmap("Set3")(range(len(route_revenue)))

# Create pie chart
plt.figure(figsize=(10, 7))
plt.pie(route_revenue, labels=route_revenue.index, autopct='%1.1f%%', 
        startangle=140, colors=colors, explode=[0.05] * len(route_revenue),
        shadow=True, wedgeprops={'edgecolor': 'black'})

plt.title('Revenue Distribution by Route Type', fontsize=14, fontweight='bold')
plt.show()
```

## Suggestions



##### Optimize Ship Deployment for Profitability:

Prioritize Bulk Carriers and Fish Carriers, as they generate the highest revenue.

Container Ships, despite high operational costs, should be used on high-demand trade routes to maximize efficiency.

##### Select Routes Strategically:

Long-haul and Transoceanic routes generate the most revenue, making them ideal for high-capacity ships.
Coastal and Short-haul routes can be leveraged for quick turnaround times with lower operating costs.

##### Improve Fuel Efficiency & Speed:

Heavy Fuel Oil (HFO) engines offer the highest speeds, making them suitable for long-distance voyages.
Investing in fuel-efficient technologies can reduce operational costs.

##### Enhance Cargo Load Efficiency:

Tankers and Fish Carriers have the highest load efficiency, meaning companies should optimize cargo capacity to improve profitability.

##### Manage Operational Costs:

Since Container Ships have the highest operational costs, optimizing fuel consumption, maintenance, and port fees is crucial to maintaining profit margins.


### Recommendations



Best Ship Type for Profitability → Bulk Carriers generate the highest revenue while keeping operational costs relatively lower.


Best Routes for Revenue → Long-haul and Transoceanic routes generate the most income, making them ideal for strategic fleet deployment.


Most Efficient Ships → Tankers travel the longest distances and have the highest load efficiency, making them crucial for fuel transport.


Fastest Engine Type → Heavy Fuel Oil (HFO) engines are the best choice for speed.


Most Expensive to Operate → Container Ships have the highest operational cost, but they also generate strong revenues.


 Business Strategy: For maximum profitability, focus on Bulk Carriers and Tankers on Long-haul routes while optimizing operational costs for Container Ships.
